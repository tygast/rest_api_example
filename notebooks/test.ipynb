### Clone Assets

```python
def asset_template(_resp, parent, old, hub):
    add_assets = []
    for item in _resp:
        template = {}
        template['ID'] = item['ID'].replace(old,hub['ID'])
        template['Name'] = item['Name']
        template['IsLocation'] = item['IsLocation']
        template['Vicinity'] = item['Vicinity'].replace(old,hub['PID']) 
        template['Icon'] = item['Icon']
        template['ParentRef'] = {'ID': parent.replace(old,hub['ID'])}
        template['ClassificationRef'] = item['ClassificationRef']
        template['RepairCenterRef'] = {'ID': item['RepairCenterRef']['ID'].replace(old,hub['ID'])}
        try:
            template['ShopRef'] = {'ID': item['ShopRef']['ID'].replace(old,hub['ID'])}
        except TypeError:
            pass
        template['TypeDetails'] = item['TypeDetails']
        add_assets.append(copy.deepcopy(template))
    return add_assets
```

```python
def asset_cloner(parent, old, hub):
    get_assets = _get(module='assets', _filter='ParentRef.ID', operator='eq', identifier=parent, top=500, skip=0)
    add_assets = asset_template(get_assets, parent, old, hub)
    for asset in add_assets:
        _resp = _post('assets',[asset])
```
