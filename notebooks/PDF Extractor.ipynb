```python
def text_extractor(path, pid):
    tags = []
    with open(path, 'rb') as f:
        pdf = PdfFileReader(f) 
        page = pdf.getPage(0)
        items = list(range(0,len(page['/Annots'])))
        for i in items:
            tag_detail = page['/Annots'][i].getObject()
            print(tag_detail)
            tag_num = tag_detail['/Contents']
            try:
                tag_prefix = page['/Annots'][i+1].getObject()['/Contents']
            except IndexError: 
                pass
            if all(str.isdigit(d) for d in tag_num[:3]) and tag_prefix.isalpha() and len(tag_num)>=4 and ' ' not in tag_num and '/' not in tag_prefix and '@' not in tag_prefix and '"' not in tag_prefix and '-' not in tag_num and '#' not in tag_num:
                tags.append(tag_prefix+'-'+tag_num)
    return tags
```

```python
tags = text_extractor(path, smi_pid_1101a)
```

```python
def assets_to_dict(page):
    keys = list(page.keys())
    check_tags = []
    for key in keys:
        for tag in tags:
            if key[-4:] in tag:
                page[key].append(tag)
                check_tags.append(tag)
    no_match = [x for x in tags if x not in check_tags]
    page['SKID'].append(no_match)
    return page
```
